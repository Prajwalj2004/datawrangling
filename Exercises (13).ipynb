### Step 1. Import the necessary libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

```python
url=r'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
```

### Step 3. Assign it to a variable called tips

```python
tips=pd.read_csv(url)
tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
tips=tips.drop(columns='Unnamed: 0')
```

```python
tips
```

### Step 5. Plot the total_bill column histogram

```python
plt.hist(tips['total_bill']) #ylabel is count
```

```python
#create histogram
ttbill=sns.distplot(tips['total_bill'])

#set labels and titles
ttbill.set(xlabel='Value',ylabel='Frequency',title='Total_bill')

#take out the right and the upper border
#sns.despine()
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'])
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'])
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],size=tips['size'])
```

```python
sns.PairGrid(tips,vars=['total_bill','tip','size'])
```

```python
sns.pairplot(tips,diag_kind='kde',kind='reg',hue='time') #when it is a single number column,it gives histogram 
#diag_kind-gives curve to the single number column
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['size'],palette='colorblind') 
#palette=muted,colorblind,deep,bright,pastel
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.bar(tips['day'],tips['total_bill'],width=0.4)
```

```python
sns.stripplot(x=tips['day'],y=tips['total_bill'],jitter=True,orient='v',hue=tips['sex'],dodge=True,
              legend='auto')
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.scatterplot(x=tips['tip'],y=tips['day'],hue=tips['sex'])
```

```python
sns.stripplot(y=tips['day'],x=tips['tip'],jitter=True,hue=tips['sex'],dodge=True)
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(y=tips['total_bill'],x=tips['day'],hue=tips['time'])
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
plt.subplot(1,2,1)
Dinner=tips[tips['time']=='Dinner']
sns.histplot(Dinner,x=Dinner['tip'])
plt.subplot(1,2,2)
Lunch=tips[tips['time']=='Lunch']
sns.histplot(Lunch,x=Lunch['tip'])
```

```python
sns.set(style='ticks')
g=sns.FacetGrid(tips,col='time')
g.map(plt.hist,'tip')
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
sns.set_theme(context='p',style='ticks',palette='deep')
g=sns.FacetGrid(tips,col='sex')
g.map(sns.scatterplot,'tip','total_bill',hue=tips['smoker'])
```

```python
sns.set(style='ticks')
g=sns.FacetGrid(tips,col='sex',hue='smoker')
g.map(plt.scatter,'tip','total_bill')
```
